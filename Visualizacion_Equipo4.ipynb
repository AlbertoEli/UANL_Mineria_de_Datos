###**2. Descripción de los datos**

a) ¿Qué tipo de datos son?



> Los datos a trabajar son tablas que contienen las características que se tienen que tomar en cuenta para categorizar el agua como potable o no potable.




b) Descripción de las columnas.



> Las columnas contienen métricas de la calidad del agua para 3276 cuerpos de agua distintos.

>Descripción de columnas:

* ph
>ph del agua.

* Dureza
> Capacidad del agua para precipitar jabón en mg/L.

* Sólidos	
>Sólidos totales disueltos en ppm (Partes por millón).

* Cloraminas
>Cantidad de Cloraminas en ppm.

* Sulfato
>Cantidad de sulfatos disueltos en mg/L.

*	Conductividad
>Conductividad eléctrica del agua en μS/cm (microSiemens por centímetro). 

*	Carbón_orgánico
>Cantidad de carbono orgánico en ppm.

*	Trihalometanos
>Cantidad de trihalometanos en μg/L (Microgramo por Litro).

*	Turbiedad
>Medida de la propiedad de emisión de luz del agua en NTU (Unidades de turbidez nefelométrica).

*	Potabilidad
>Indica si el agua es segura para el consumo humano.

####Estadística básica de los datos

Aisgnación de datos a variable 

```python
import pandas as pd
import numpy as np
import json
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("water_potability.csv")
```

Ya que se tienen los datos asignados a la variable "df", se procede a visualizar los datos.

```python
#Visualización de los datos 
df
```

Al observar los datos de la base de datos notamos que cuentan con datos nulos, por lo que procedemos a eliminarlos para evitar errores al momento del cálculo de medidas estadísticas. Al igual, se cambian los titulos de las columnas a español para un manejo más sencillo

```python
df_drop = df.dropna() #eliminamos las filas donde se tiene al menos un valor nulo

df_drop.insert(0,'id',df_drop.index + 1,False) #insertamos columna de id para trabajar los datos más adelante
df_drop = df_drop.rename(columns = {
    'Hardness' : 'Dureza',
    'Solids' : 'Sólidos',
    'Chloramines' : 'Cloraminas',
    'Sulfate' : 'Sulfato',
    'Conductivity' : 'Conductividad',
    'Organic_carbon' : 'Carbón_orgánico',
    'Trihalomethanes' : 'Trihalometanos',
    'Turbidity':'Turbiedad',
    'Potability':'Potabilidad'
    
})
print(" ")
print("Primeros 5 datos de la Base de Datos: ")
df_drop.head()
```

Ya que se encuentran los datos sin ningun faltante, se procede a buscar las medidas estadísticas

1. Suma.

```python
#Creando una lista para guardar las sumas
sumas=[]

#Calculando las sumas para cada columna a excepción del id
for i in range(0,len(df_drop.columns)-1):
  sumas.append(df_drop[df_drop.columns[i+1]].sum())

#Creando una lista con los nombres de las columnas 
a=list(df_drop.columns)
for i in range(0,len(sumas)):
 print("La suma de la columna", a[i+1], "es: ", sumas[i])
```

2. Promedio.

```python
#Creando una lista para guardar los promedios
promedios=[]

#Calculando las sumas para cada columna a excepción del id
for i in range(0,len(df_drop.columns)-1):
  promedios.append(df_drop[df_drop.columns[i+1]].mean())

#Creando una lista con los nombres de las columnas 
a=list(df_drop.columns)

for i in range(0,len(promedios)):
 print("El promedio de la columna", a[i+1], "es: ", promedios[i])
```

3. Suma acumulada de la columna fila por fila.

```python
#Creando una lista para guardar los promedios
cumsums=[]

#Calculando las sumas acumuladas para cada columna a excepción del id
for i in range(0,len(df_drop.columns)-1):
  cumsums.append(df_drop[df_drop.columns[i+1]].cumsum())

#Creando una lista con los nombres de las columnas 
a=list(df_drop.columns)

for i in range(0,len(cumsums)):
 print("La suma acumulada de la columna", a[i+1], "es: ")
 print(cumsums[i])
```

4. Resumen estadístico de cada columna.

```python
#Creando una lista para guardar los resúmenes de cada columna
resumen=[]

#Obteniendo el resumen para cada columna a excepción del id
for i in range(0,len(df_drop.columns)-1):
  resumen.append(df_drop[df_drop.columns[i+1]].describe())

#Creando una lista con los nombres de las columnas 
a=list(df_drop.columns)

for i in range(0,len(resumen)):
 print("El resumen de la columna", a[i+1], "es: ")
 print(resumen[i])
 print(" ")
```

5. Contar elementos de las columnas.

Anteriormente se eliminaron las filas que contenían al menos 1 elemento nulo, por lo tanto, la función count() nos arrojará el mismo número para todas las columnas. Con utilizar dicha función en 1 columna es suficiente.

```python
print("Cada columna tiene", df_drop['Potabilidad'].count(), "elementos.")
```

6. Mínimo y máximo.



*   Mínimo



```python
#Creando una lista para guardar los mínimos y máximos
mins=[]
maxs=[]

#Calculando los mínimos y máximos para cada columna a excepción del id
for i in range(0,len(df_drop.columns)-1):
  mins.append(df_drop[df_drop.columns[i+1]].min())
  maxs.append(df_drop[df_drop.columns[i+1]].max())

#Creando una lista con los nombres de las columnas 
a=list(df_drop.columns)
for i in range(0,len(mins)):
 print("El mínimo de la columna", a[i+1], "es:", mins[i], "y su máximo es:", maxs[i])
```

7. Mediana, varianza y desviación estándar.



* Mediana



```python
#Creando una lista para guardar las medianas
medianas=[]

#Calculando las medianas para cada columna a excepción del id
for i in range(0,len(df_drop.columns)-1):
  medianas.append(df_drop[df_drop.columns[i+1]].median())

#Creando una lista con los nombres de las columnas 
a=list(df_drop.columns)
for i in range(0,len(medianas)):
 print("La mediana de la columna", a[i+1], "es: ", medianas[i])
```

* Varianza

```python
#Creando una lista para guardar las varianzas
varianzas=[]

#Calculando las varianzas para cada columna a excepcion del id
for i in range(0,len(df_drop.columns)-1):
  varianzas.append(df_drop[df_drop.columns[i+1]].var())

#Creando una lista con los nombres de las columnas 
a=list(df_drop.columns)
for i in range(0,len(varianzas)):
 print("La varianza de la columna", a[i+1], "es: ", varianzas[i])
```



* Desviación estándar





```python
#Creando una lista para guardar las desviaciones estándar
de=[]

#Calculando las varianzas para cada columna a excepcion del id
for i in range(0,len(df_drop.columns)-1):
  de.append(df_drop[df_drop.columns[i+1]].std())

#Creando una lista con los nombres de las columnas 
a=list(df_drop.columns)
for i in range(0,len(de)):
 print("La desviación estándar de la columna", a[i+1], "es: ", de[i])
```

8. Valor de asimetría en los datos.

```python
#Creando una lista para guardar los valores de asimetría
asi=[]

#Calculando las varianzas para cada columna a excepcion del id
for i in range(0,len(df_drop.columns)-1):
  asi.append(df_drop[df_drop.columns[i+1]].skew())

#Creando una lista con los nombres de las columnas 
a=list(df_drop.columns)
for i in range(0,len(asi)):
 print("La asimetría de la columna", a[i+1], "es: ", asi[i])
```

9. Curtosis

```python
#Creando una lista para guardar los valores de curtosis
cur=[]

#Calculando las varianzas para cada columna a excepcion del id
for i in range(0,len(df_drop.columns)-1):
  cur.append(df_drop[df_drop.columns[i+1]].kurt())

#Creando una lista con los nombres de las columnas 
a=list(df_drop.columns)
for i in range(0,len(cur)):
 print("La curtosis de la columna", a[i+1], "es: ", cur[i])
```
